```python
# --- Tabla 10: Ranking de Crecimiento de Casos por Departamento (2015 vs. 2023) ---
# Se incluyen los casos de 2015, 2023 y la tasa de crecimiento calculada.

datos_tabla_10 = {
    'Ranking': list(range(1, 19)),
    'Departamento': [
        'ITAPUA', 'BOQUERON', 'CAAGUAZU', 'SAN PEDRO', 'PTE. HAYES', 'ALTO PARANA',
        'CORDILLERA', 'ALTO PARAGUAY', 'PARAGUARI', 'GUAIRA', 'CAPITAL',
        'MISIONES', 'CENTRAL', 'CANINDEYU', 'AMAMBAY', 'CONCEPCION', 'CAAZAPA', 'ÑEEMBUCU'
    ],
    'Casos en 2015': [1189, 18, 878, 693, 157, 1847, 721, 33, 673, 623, 2128, 504, 4204, 311, 420, 325, 336, 218],
    'Casos en 2023': [2145, 32, 1567, 1245, 281, 3215, 1298, 59, 1208, 1119, 3467, 904, 6862, 558, 753, 582, 602, 391]
}

# Create the DataFrame first
df_tabla10 = pd.DataFrame(datos_tabla_10)


# --- (Opcional) Calcular la columna de crecimiento directamente en el DF ---
# La Tasa de Crecimiento Anual Compuesta (TCAC) se calcula sobre 8 períodos (2015 a 2023).
# TCAC = ((Valor Final / Valor Inicial) ^ (1 / N_Periodos)) - 1
df_tabla10['Crecimiento Anual (TCAC)'] = \
    (np.power(df_tabla10['Casos en 2023'] / df_tabla10['Casos en 2015'], 1/8) - 1) * 100
# Formatear la columna como un string de porcentaje
df_tabla10['Crecimiento Anual (TCAC)'] = df_tabla10['Crecimiento Anual (TCAC)'].round(2).astype(str) + '%'

print("--- Tabla 10: Ranking de Crecimiento de Casos por Departamento (2015 vs. 2023) ---")
# Ordenar por el valor numérico antes de convertirlo a string para la visualización.
df_tabla10_sorted = df_tabla10.sort_values(by='Crecimiento Anual (TCAC)', ascending=False).reset_index(drop=True)
print(df_tabla10_sorted.to_string(index=False))
```

```python
# prompt: crear un gráfico de barras con TCAC de la tabla previa

# Asegúrate de que la columna TCAC sea numérica para graficar
# Primero, limpia la columna TCAC eliminando el '%' y convirtiéndola a float
df_tabla10['Crecimiento Anual (TCAC_numeric)'] = df_tabla10['Crecimiento Anual (TCAC)'].str.replace('%', '').astype(float)

# Ordenar por la columna numérica de TCAC para el gráfico
df_tabla10_sorted_for_plot = df_tabla10.sort_values(by='Crecimiento Anual (TCAC_numeric)', ascending=False).reset_index(drop=True)

# Figura 10 - Gráfico de barras para el Ranking de Crecimiento (TCAC) por Departamento
plt.figure(figsize=(14, 10))

# Creamos el gráfico de barras
plt.bar(df_tabla10_sorted_for_plot['Departamento'], df_tabla10_sorted_for_plot['Crecimiento Anual (TCAC_numeric)'], color='lightblue')

# Añadimos títulos y etiquetas
plt.title('Ranking de Crecimiento Anual Compuesto (TCAC) de Casos por Departamento (2015-2023)')
plt.xlabel('Departamento')
plt.ylabel('TCAC (%)')

# Rotamos las etiquetas del eje x para una mejor lectura
plt.xticks(rotation=90)

# Mostramos el valor de TCAC encima de cada barra
for i, row in df_tabla10_sorted_for_plot.iterrows():
    plt.text(i, row['Crecimiento Anual (TCAC_numeric)'], f"{row['Crecimiento Anual (TCAC)']}", ha='center', va='bottom')

plt.tight_layout() # Ajusta el layout
plt.show()
```
